```python
import pandas as pd
import matplotlib.pyplot as plt

# Đường dẫn file
file_path = r"D:\LDA\DL\_DCS_Items__202512051333.csv"

# Chỉ đọc các cột cần thiết
use_cols = [
    "CronTime",
    "B1_FT1151_DACA_PV__Value",
    "B2_FT1151_DACA_PV__Value",
    "B1_FT1151_DIVA_OUT__Value",
    "B2_FT1151_DIVA_OUT__Value"
]

# Đọc theo từng phần để tránh lỗi MemoryError
chunks = []
for chunk in pd.read_csv(file_path, usecols=use_cols, chunksize=200_000):
    chunks.append(chunk)

# Gộp các chunk lại
df = pd.concat(chunks, ignore_index=True)

# Chuyển timestamp
df['CronTime'] = pd.to_datetime(df['CronTime'])

# ======================
# Công suất lò 1 & lò 2
# ======================

fig, axes = plt.subplots(2, 1, figsize=(14, 10), sharex=True)

# 1️⃣ Công suất lò 1
axes[0].plot(df['CronTime'], df['B1_FT1151_DACA_PV__Value'], color='blue')
axes[0].set_ylabel("Công suất (tấn/h)")
axes[0].set_title("Công suất theo thời gian")
axes[0].grid(True)

# 2️⃣ Công suất lò 1
axes[1].plot(df['CronTime'], df['B2_FT1151_DACA_PV__Value'], color='orange')
axes[1].set_ylabel("Công suất (tấn/h)")
axes[1].set_title("Công suất theo thời gian")
axes[1].grid(True)

plt.tight_layout()
plt.show()

# ======================
# Công suất ↔ Tiêu hao
# ======================
plt.figure(figsize=(8, 6))
plt.scatter(df['B1_FT1151_DACA_PV__Value'], df['B1_FT1151_DIVA_OUT__Value'], s=5)
plt.yscale("log")

plt.title("Quan hệ giữa Công suất và Tiêu hao")
plt.xlabel("Công suất lò 1 (tấn/h)")
plt.ylabel("Tiêu hao trên 1 tấn hơi cao áp")
plt.grid(True)
plt.tight_layout()
plt.show()

# ======================
plt.figure(figsize=(8, 6))
plt.scatter(df['B2_FT1151_DACA_PV__Value'], df['B2_FT1151_DIVA_OUT__Value'], s=5)
plt.yscale("log")

plt.title("Quan hệ giữa Công suất và Tiêu hao")
plt.xlabel("Công suất lò 2 (tấn/h)")
plt.ylabel("Tiêu hao trên 1 tấn hơi cao áp")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# File CSV
file_path = r"D:\LDA\DL\_DCS_Items__202512051333.csv"

# Các cột cần dùng
use_cols = [
    "CronTime",
    "B1_TE1251_DACA_PV__Value",   # Nhiệt độ nước cấp
    "B1_Z_BEDT_DACA_PV__Value",   # Nhiệt độ buồng đốt
    "B1_TE1212_DACA_PV__Value",   # Nhiệt độ hơi quá nhiệt
    "B1_TE1111_DACA_PV__Value",   # Đuôi lò TE1111
    "B1_TE1112_DACA_PV__Value"    # Đuôi lò TE1112
]

# Đọc dữ liệu theo chunks để tránh lỗi bộ nhớ
chunks = []
for chunk in pd.read_csv(file_path, usecols=use_cols, chunksize=200_000):
    chunks.append(chunk)

df = pd.concat(chunks, ignore_index=True)

# Convert thời gian
df["CronTime"] = pd.to_datetime(df["CronTime"])
df = df.sort_values("CronTime")

# OPTIONAL: smoothing để giảm nhiễu
window = 50

df_smooth = df.copy()
num_cols = df.select_dtypes(include='number').columns

df_smooth[num_cols] = df[num_cols].rolling(window=window, min_periods=1).mean()


# ----------------------
# Vẽ biểu đồ
# ----------------------
plt.figure(figsize=(16, 10))

# --- SUBPLOT 1: Nhiệt độ chính (nước cấp – buồng đốt – hơi quá nhiệt)
plt.subplot(2, 1, 1)
plt.plot(df["CronTime"], df_smooth["B1_TE1251_DACA_PV__Value"], label="Nhiệt độ nước cấp")
plt.plot(df["CronTime"], df_smooth["B1_Z_BEDT_DACA_PV__Value"], label="Nhiệt độ buồng đốt")
plt.plot(df["CronTime"], df_smooth["B1_TE1212_DACA_PV__Value"], label="Nhiệt độ hơi quá nhiệt")

plt.title("Nhiệt độ nước cấp – buồng đốt – hơi quá nhiệt")
plt.xlabel("Thời gian")
plt.ylabel("Nhiệt độ (°C)")
plt.legend()
plt.grid(True)

# --- SUBPLOT 2: Nhiệt độ đuôi lò (2 cảm biến TE1111 – TE1112)
plt.subplot(2, 1, 2)
plt.plot(df["CronTime"], df_smooth["B1_TE1111_DACA_PV__Value"], label="TE1111 (đuôi lò)")
plt.plot(df["CronTime"], df_smooth["B1_TE1112_DACA_PV__Value"], label="TE1112 (đuôi lò)")

plt.title("Nhiệt độ đuôi lò TE1111 – TE1112")
plt.xlabel("Thời gian")
plt.ylabel("Nhiệt độ (°C)")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

```
